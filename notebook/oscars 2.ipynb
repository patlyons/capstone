```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
```

```python
#oscars 1998
URL = 'https://www.oscars.org/oscars/ceremonies/1998' # /{year}

response = requests.get(URL)

soup=bs(response.text)

oscar_soup = soup.findAll('div', attrs={'class':'view-grouping'})
```

```python
oscar_98 = pd.DataFrame().from_dict({x.find('h2').text: [y.text for y in x.findAll('h4')] for x in oscar_soup}, orient = 'index').T.melt(var_name='category', value_name='nominee')

```
