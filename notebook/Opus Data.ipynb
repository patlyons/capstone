# how much have movie budgets increased from 2006 to 2018

```python
y2006 = pd.DataFrame(moviedata.loc[moviedata.production_year ==2006])
```

```python
y2006
```

```python
(y2006.production_budget.sum())/len(y2006.production_budget)
```

```python
y2006.production_budget.mean()
```

```python
y2018 = pd.DataFrame(moviedata.loc[moviedata.production_year ==2018])
round((y2018.production_budget.sum())/len(y2018.production_budget),2)
```

```python
round((68071794.87 - 44693750.0),2)
```

```python
round((68071794.87/44693750.0), 2)*100
```

```python
avg_2006_budget = (y2006.production_budget.sum())/len(y2006.production_budget)
y2018 = pd.DataFrame(moviedata.loc[moviedata.production_year ==2018])
avg_2018_budget = (y2018.production_budget.sum())/len(y2018.production_budget)
avg_difference = avg_2018_budget - avg_2006_budget
twelve_year_percent_increase = round((avg_2018_budget / avg_2006_budget ),2)*100
print(str('twelve year increased budget percent: '), twelve_year_percent_increase, str('%') )
```
