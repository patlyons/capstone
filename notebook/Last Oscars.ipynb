```python
#oscars 1998
URL = 'https://www.oscars.org/oscars/ceremonies/1998' # /{year}

response = requests.get(URL)

soup=bs(response.text)

oscar_soup = soup.findAll('div', attrs={'class':'view-grouping'})
```

```python
oscar_soup2 = soup.findAll('div', attrs={'class':'view-grouping'})

oscar_soup2
```

```python
dict = {}
for group in oscar_soup2: 
    try: 
        
        dict[group.find('h2').text] = [entry.text.strip() for entry in group.findAll('div', attrs={'class':'views-field-title'})]
    except:
        pass
```

```python
oscars_98 = pd.DataFrame.from_dict(dict, orient = 'index').T
oscars_98['year'] = 1998
oscars98_df = oscars_98[['Writing (Screenplay Written Directly for the Screen)', 'year']]
oscars98_df
#oscars98_df.drop[1:4]
```
