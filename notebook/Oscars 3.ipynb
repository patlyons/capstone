```python
dict = {}
for group in oscar_soup3: 
    try: 
        
        dict[group.find('h2').text] = [entry.text.strip() for entry in group.findAll('div', attrs={'class':'views-field-title'})]
    except:
        pass
    #print(group.find('div', attrs={'class':'views-field views-field-title'}).find('span', {'class' : 'field-content' }).text)
    #print([entry.text for entry in group.findAll('div', attrs={'class':'views-field-title'})])
    
    
    
```
