```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np

#URL for list of all oscar nominees over all years
URL = 'https://www.oscars.org/oscars/ceremonies' # /{year}
#get response
response = requests.get(URL)
#make it into a soup
soup=bs(response.text)

#create soup for nominees
oscar_soup = soup.findAll('div', attrs={'class':'view-grouping'})

oscar_2022 = pd.DataFrame().from_dict({x.find('h2').text: [y.text for y in x.findAll('h4')] for x in oscar_soup},
                                      orient = 'index').T.melt(var_name='category', value_name='nominee')
oscar_2022 = oscar_2022.dropna()
oscar_2022['year'] = 2022
oscar_2022
```
